# ⚙️🤖🦾🛠️ Atuo ML

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE

# 1. Prepare your data (Encoding and dropping non-predictive columns)
X = pd.get_dummies(df.drop(['Churn Label', 'Customer Status', 'Country', 'State', 'Quarter'], axis=1), drop_first=True)
y = LabelEncoder().fit_transform(df['Churn Label'])

# 2. Initial Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 3. Scaling (LazyPredict can handle some scaling, but doing it manually is safer)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Apply SMOTE to training data only
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

# 5. Run LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train_res, X_test_scaled, y_train_res, y_test)

# Display the performance table
print(models)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import AdaBoostClassifier, VotingClassifier
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

# --- PREPARE DATA ---
X = pd.get_dummies(df.drop(['Churn Label', 'Customer Status', 'Country', 'State', 'Quarter'], axis=1), drop_first=True)
y = LabelEncoder().fit_transform(df['Churn Label'])

X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# --- DEFINE PARAMETER GRIDS ---
grids = {
    'AdaBoost': (AdaBoostClassifier(random_state=42),
                 {'n_estimators': [50, 100, 200,300,400,450], 'learning_rate': [0.001,0.01, 0.1, 1.0]}),
    'LGBM': (LGBMClassifier(random_state=42, verbosity=-1),
             {'n_estimators': [100, 200,250,299,400], 'learning_rate': [0.001,0.01, 0.1, 1.0], 'num_leaves': [20,31, 50,70,80]}),
    'DecisionTree': (DecisionTreeClassifier(random_state=42),
                     {'max_depth': [None, 10, 20,30,40,50], 'min_samples_split': [2, 5, 10,15,20]})
}

best_estimators = {}

# --- MANUAL FOLD + HALVING SEARCH LOOP ---
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scaler = StandardScaler()
smote = SMOTE(random_state=42)

for name, (model_obj, param_grid) in grids.items():
    print(f"\n--- Tuning {name} ---")


    X_t_scaled = scaler.fit_transform(X_train_full)
    X_t_res, y_t_res = smote.fit_resample(X_t_scaled, y_train_full)

    halving = HalvingGridSearchCV(model_obj, param_grid, cv=skf, factor=2, scoring='f1')
    halving.fit(X_t_res, y_t_res)

    print(f"Best Params for {name}: {halving.best_params_}")
    best_estimators[name] = halving.best_estimator_

# --- 2. VOTING CLASSIFIER ---
voting_clf = VotingClassifier(
    estimators=[(name, clf) for name, clf in best_estimators.items()],
    voting='soft' # 'soft' uses probabilities for better voting weight
)

voting_clf.fit(X_t_res, y_t_res)
best_estimators['Voting'] = voting_clf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (f1_score, confusion_matrix, roc_curve,
                             auc, precision_recall_curve, classification_report, accuracy_score)

# --- 1. PREPARE DATA FOR COMPARISON ---
results_f1 = {}
results_acc = {}
X_test_scaled = scaler.transform(X_test_full)

for name, clf in best_estimators.items():
    y_pred = clf.predict(X_test_scaled)
    results_f1[name] = f1_score(y_test_full, y_pred)
    results_acc[name] = accuracy_score(y_test_full, y_pred)

# --- 2. ENHANCED BAR CHART (F1-SCORE WITH PERCENTAGES) ---
plt.figure(figsize=(10, 6))
names = list(results_f1.keys())
values = list(results_f1.values())

bars = plt.bar(names, values, color=sns.color_palette("viridis", len(names)))

# Adding percentage labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02,
             f'{yval*100:.1f}%', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('Model Comparison: F1-Score (%)', fontsize=15)
plt.ylim(0, 1.15)
plt.ylabel('F1-Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('f1_comparison.png')
plt.show()

# --- 3. NORMALIZED CONFUSION MATRIX (PERCENTAGES) ---
y_pred_voting = voting_clf.predict(X_test_scaled)
cm = confusion_matrix(y_test_full, y_pred_voting)
cm_perc = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] # Normalize by row

plt.figure(figsize=(8, 6))
# Annotate with both count and percentage
labels = (np.array([f"{count}\n({perc:.1%})"
          for count, perc in zip(cm.flatten(), cm_perc.flatten())])).reshape(2,2)

sns.heatmap(cm_perc, annot=labels, fmt="", cmap='Blues', cbar=False)
plt.title('Voting Classifier: Normalized Confusion Matrix', fontsize=14)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.savefig('confusion_matrix.png')
plt.show()

# --- 4. ROC CURVE COMPARISON ---
plt.figure(figsize=(10, 7))
for name, clf in best_estimators.items():
    if hasattr(clf, "predict_proba"):
        y_prob = clf.predict_proba(X_test_scaled)[:, 1]
    else:
        y_prob = clf.predict(X_test_scaled)

    fpr, tpr, _ = roc_curve(y_test_full, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.3f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.title('ROC Curve Comparison', fontsize=15)
plt.legend(loc="lower right")
plt.savefig('roc_curve.png')
plt.show()

# --- 5. FEATURE IMPORTANCE (FOR TOP MODEL: LGBM) ---
lgbm_model = best_estimators['LGBM']
importances = lgbm_model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False).head(10)

plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='magma')
plt.title('Top 10 Drivers of Churn (LGBM)', fontsize=15)
plt.savefig('feature_importance.png')
plt.show()
```
